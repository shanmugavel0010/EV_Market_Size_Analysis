```python

#Perform clustering analysis to identify hotspots of EV adoption. 

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Step 1 - Prepare Data
city_counts_df = city_counts.reset_index()
city_counts_df.columns = ['City', 'EV_Count']
city_counts_df = city_counts_df.tail(20)

# Step 2 - Apply KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
city_counts_df['Cluster'] = kmeans.fit_predict(city_counts_df[['EV_Count']])

# Step 3 - Visualize Clusters
plt.figure(figsize=(6,4))
colors = ['red', 'yellow', 'blue']
for cluster in range(3):
    cluster_data = city_counts_df[city_counts_df['Cluster'] == cluster]
    plt.scatter(cluster_data['City'], 
                cluster_data['EV_Count'],
                c=colors[cluster],
                label=f'Cluster {cluster}',
                s=100)
plt.xticks(rotation=90)
plt.xlabel('City')
plt.ylabel('EV Count')
plt.title('EV Adoption Clusters by City')
plt.legend()
plt.show()
```

```python
#predict future EV registration trends based on historical data using linear regression or time-series forecasting models.
from sklearn.linear_model import LinearRegression
import numpy as np

ev_by_year = df.groupby('Model Year')['VIN (1-10)'].count()

X = ev_by_year.index.values.reshape(-1,1)
y = ev_by_year.values

model = LinearRegression()
model.fit(X, y)

future_years = np.array([2024,2025,2026,2027,2028]).reshape(-1,1)
predictions = model.predict(future_years)

for year, pred in zip([2024,2025,2026,2027,2028], predictions):
    print(f"{year} → {int(pred):,} EVs predicted")


plt.figure(figsize=(10,5))
plt.plot(ev_by_year.index, ev_by_year.values, 
         marker='o', label='Actual Data', color='blue')
plt.plot([2024,2025,2026,2027,2028], predictions, 
         marker='o', label='Predicted', 
         color='red', linestyle='--')
plt.xlabel('Year')
plt.ylabel('Number of EVs')
plt.title('EV Registration Forecast')
plt.legend()
plt.show()
```
